```python
def preprocess_data(df):
    # Strip and lower string columns where applicable
    if "User ID" in df.columns:
        df["User ID"] = df["User ID"].fillna('').astype(str).str.strip().str.lower()
    
    if "Quiz Topic" in df.columns:
        df["Quiz Topic"] = df["Quiz Topic"].fillna('').astype(str).str.strip().str.lower()
    
    if "Quiz Title" in df.columns:
        df["Quiz Title"] = df["Quiz Title"].fillna('').astype(str).str.strip().str.lower()
    
    # Handle percentage strings in specific columns
    if "Score" in df.columns:
        df["Score"] = df["Score"].replace('%', '', regex=True).astype(float)
    
    if "Accuracy" in df.columns:
        df["Accuracy"] = df["Accuracy"].replace('%', '', regex=True).astype(float)
    
    # Convert duration from string (e.g., '15:00') to numeric minutes
    if "Duration" in df.columns:
        df["Duration"] = df["Duration"].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]) if isinstance(x, str) else x)

    return df

# Preprocess both datasets
history_data = preprocess_data(history_data)
current_data = preprocess_data(current_data)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Function to preprocess the data
def preprocess_data(df):
    # Strip whitespaces and convert IDs to lowercase (convert to string first)
    df["Quiz ID"] = df["Quiz ID"].astype(str).str.strip().str.lower()
    df["User ID"] = df["User ID"].astype(str).str.strip().str.lower()
    
    # Ensure 'Topic' column exists and is cleaned (convert to lowercase)
    if 'Quiz Topic' in df.columns:
        df["Topic"] = df["Quiz Topic"].astype(str).str.strip().str.lower()
    elif 'Topic' in df.columns:
        df["Topic"] = df["Topic"].astype(str).str.strip().str.lower()
    
    # Split topics if they are concatenated (assuming topics are separated by 'and' or similar)
    df["Topic"] = df["Topic"].apply(lambda x: x.split('and'))  # Split by 'and' if that's the delimiter
    
    # Flatten the lists (in case there are multiple topics for each row)
    df = df.explode("Topic").reset_index(drop=True)
    
    # Remove duplicates and create an index for each unique topic
    unique_topics = df["Topic"].unique()
    topic_mapping = {topic: idx for idx, topic in enumerate(unique_topics)}
    
    # Replace 'Topic' values with their corresponding indices
    df["Topic_index"] = df["Topic"].map(topic_mapping)
    
    # Create a separate mapping for index to topic
    index_to_topic = {v: k for k, v in topic_mapping.items()}
    
    # Convert percentage columns (Score, Accuracy) to numeric by removing '%'
    def convert_percentage_to_numeric(val):
        if isinstance(val, str) and '%' in val:
            return float(val.replace('%', '')) / 100
        return pd.to_numeric(val, errors='coerce')  # Convert to numeric safely
    
    # Apply the function to 'Score' and 'Accuracy' columns
    if 'Score' in df.columns:
        df['Score'] = df['Score'].apply(convert_percentage_to_numeric)
    if 'Accuracy' in df.columns:
        df['Accuracy'] = df['Accuracy'].apply(convert_percentage_to_numeric)
    
    # Ensure 'Score' and 'Accuracy' are numeric
    df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
    df['Accuracy'] = pd.to_numeric(df['Accuracy'], errors='coerce')
    
    # Drop rows where 'Score' or 'Accuracy' is NaN after conversion
    df = df.dropna(subset=['Score', 'Accuracy'])
    
    return df, index_to_topic

# Load datasets
history_path = "history.csv"  # Path to history.csv
current_path = "current.csv"  # Path to current.csv

history_data = pd.read_csv(history_path)
current_data = pd.read_csv(current_path)

# Apply preprocessing
history_data, history_index_to_topic = preprocess_data(history_data)
current_data, current_index_to_topic = preprocess_data(current_data)

# Combine the two mappings (if you want them to be the same across datasets)
index_to_topic = {**history_index_to_topic, **current_index_to_topic}

# Step 3: Remove NaN values from both datasets
history_data = history_data.dropna()
current_data = current_data.dropna()

# Step 4: Feature Engineering
# Historical performance metrics
history_performance = (
    history_data.groupby("Topic_index")
    .agg({"Score": "mean", "Accuracy": "mean"})
    .reset_index()
)

# Current performance metrics
current_performance = (
    current_data.groupby("Topic_index")
    .agg({"Correct Answers": "sum", "Total Questions": "sum"})
    .reset_index()
)
current_performance["Accuracy"] = (
    current_performance["Correct Answers"] / current_performance["Total Questions"]
)

# Combine historical and current data
performance_data = pd.merge(
    history_performance, current_performance, on="Topic_index", how="outer", suffixes=("_history", "_current")
)

# Step 5: Normalize the data
scaler = StandardScaler()
performance_data_scaled = scaler.fit_transform(
    performance_data[["Score", "Accuracy_history", "Accuracy_current", "Topic_index"]]
)

# Step 6: Impute NaN values (if any) after scaling
imputer = SimpleImputer(strategy="mean")
performance_data_scaled = imputer.fit_transform(performance_data_scaled)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
performance_data["Cluster"] = kmeans.fit_predict(performance_data_scaled)

# Step 7: Weakness and Strength Analysis
# Weaknesses: Low accuracy and scores
weak_topics = performance_data[(
    performance_data["Accuracy_current"] < 0.5) | (performance_data["Score"] < 0.5)
]["Topic_index"].tolist()

# Strengths: High accuracy and scores
strong_topics = performance_data[(
    performance_data["Accuracy_current"] > 0.8) & (performance_data["Score"] > 0.8)
]["Topic_index"].tolist()

# Step 8: Recommendations
def recommend_topics(cluster_data, weakness_list):
    recommendations = cluster_data[cluster_data["Topic_index"].isin(weakness_list)]
    return recommendations[["Topic_index", "Accuracy_current"]]

recommendation_list = recommend_topics(performance_data, weak_topics)

# Step 9: Visualization
# Radar chart for performance
def plot_radar_chart(data, student_id):
    categories = ["Score", "Accuracy_history", "Accuracy_current"]
    values = data.mean()[categories].tolist()
    values += values[:1]  # Repeat the first value to close the radar chart

   

```

```python
import pandas as pd

# Sample data
data = {
    'Topic': [
        'Body Fluids and Circulation', 
        'Human Health and Disease', 
        'Human Reproduction', 
        'Microbes in Human Welfare', 
        'Reproductive Health', 
        'Respiration and Gas Exchange'
    ],
    'Accuracy': [
        0.45,  # Low accuracy
        0.38,  # Low accuracy
        0.72,  # Moderate accuracy
        0.83,  # High accuracy
        0.55,  # Moderate accuracy
        0.91   # Excellent accuracy
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create index to topic mapping
index_to_topic = {idx: topic for idx, topic in enumerate(df['Topic'])}

# Function to generate recommendations based on accuracy
def generate_recommendations(df):
    weak_topics = []
    strong_topics = []
    recommendations = []
    
    for index, row in df.iterrows():
        topic = row['Topic']
        accuracy = row['Accuracy']
        
        if accuracy < 0.5:
            weak_topics.append(index)
            recommendation = f"{topic}: Needs improvement. Consider revisiting key concepts for better understanding."
        elif 0.75 <= accuracy < 0.9:
            strong_topics.append(index)
            recommendation = f"{topic}: Decent performance. Strengthen understanding by reviewing key concepts for better results."
        elif accuracy >= 0.9:
            strong_topics.append(index)
            recommendation = f"{topic}: Excellent performance. Continue exploring advanced topics for deeper insights."
        else:
            recommendation = f"{topic}: Struggling with this topic. Review foundational concepts to increase retention."
        
        recommendations.append(recommendation)
    
    return weak_topics, strong_topics, recommendations

# Generate recommendations
weak_topics, strong_topics, recommendations = generate_recommendations(df)

# Print the weak and strong topics and recommendations in the required format
print("Weak Topics:", [index_to_topic[idx] for idx in weak_topics])
print("Strong Topics:", [index_to_topic[idx] for idx in strong_topics])
print("\nRecommendations for Improvement:")
for rec in recommendations:
    print(rec)
```
